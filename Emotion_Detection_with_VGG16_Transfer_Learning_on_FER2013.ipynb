```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
from collections import Counter
import cv2
from google.colab import drive, files # Import for Google Colab and file upload

# --- 1. Configuration and Data Loading ---

# Mount Google Drive (REQUIRED for Colab if you want to save the model in Drive)
print("Mounting Google Drive...")
try:
    drive.mount('/content/drive')
    print("Google Drive mounted.")
    MODEL_SAVE_PATH = '/content/drive/MyDrive/emotion_detection_vgg16_best_model.h5' # Path for saving model in Drive
except Exception as e:
    print(f"Could not mount Google Drive: {e}. Model will be saved to local Colab storage.")
    MODEL_SAVE_PATH = 'emotion_detection_vgg16_best_model.h5' # Fallback to local storage

# Define path for the dataset (will be downloaded to this location)
DATA_ROOT_DIR = '/content/fer2013_dataset' # This is where the 'train' and 'test' folders will be
TRAIN_DIR = os.path.join(DATA_ROOT_DIR, 'train')
TEST_DIR = os.path.join(DATA_ROOT_DIR, 'test') # The 'test' folder contains both public and private test sets

# Image dimensions
IMG_WIDTH, IMG_HEIGHT = 48, 48
NUM_CLASSES = 7
EMOTION_LABELS = {0: 'Angry', 1: 'Disgust', 2: 'Fear', 3: 'Happy', 4: 'Sad', 5: 'Surprise', 6: 'Neutral'}

# --- 2. Download FER2013 Dataset from Kaggle ---

dataset_downloaded_successfully = False

print("\n--- Downloading FER2013 dataset from Kaggle ---")
print("Please upload your kaggle.json file when prompted.")
print("You can get your kaggle.json from Kaggle.com -> My Account -> Create New API Token.")

# Create .kaggle directory
os.makedirs(os.path.expanduser('~/.kaggle'), exist_ok=True)

# Upload kaggle.json
try:
    uploaded = files.upload()
    if not uploaded:
        print("No kaggle.json file uploaded. Cannot proceed with dataset download.")
    else:
        for fn in uploaded.keys():
            print('User uploaded file "{name}" with length {length} bytes'.format(
                name=fn, length=len(uploaded[fn])))
        # Move kaggle.json to the correct directory
        with open(os.path.expanduser('~/.kaggle/kaggle.json'), 'wb') as f:
            f.write(uploaded[fn])
        os.chmod(os.path.expanduser('~/.kaggle/kaggle.json'), 0o600) # Set permissions
        print("kaggle.json uploaded and configured.")

        # Download the dataset
        print("Attempting to download the dataset...")
        os.makedirs(DATA_ROOT_DIR, exist_ok=True)
        import subprocess
        try:
            # The dataset ID for fer2013 is 'msambare/fer2013'
            subprocess.run(['kaggle', 'datasets', 'download', '-d', 'msambare/fer2013', '-p', DATA_ROOT_DIR, '--unzip'], check=True)
            print("Dataset downloaded and unzipped successfully.")
            if os.path.exists(TRAIN_DIR) and os.path.exists(TEST_DIR):
                print(f"Found 'train' directory at: {TRAIN_DIR}")
                print(f"Found 'test' directory at: {TEST_DIR}")
                dataset_downloaded_successfully = True
            else:
                print(f"Error: Expected 'train' and 'test' directories inside {DATA_ROOT_DIR}, but they were not found.")
                print(f"Contents of {DATA_ROOT_DIR}: {os.listdir(DATA_ROOT_DIR)}")
        except subprocess.CalledProcessError as e:
            print(f"Error downloading dataset: {e}")
            print("Please check your Kaggle API token and internet connection.")
        except Exception as e:
            print(f"An unexpected error occurred during dataset download: {e}")

except Exception as e:
    print(f"Error during kaggle.json upload or configuration: {e}")
    print("Please ensure you have uploaded your kaggle.json file and it's valid.")

# --- Conditional Execution for Preprocessing and Training ---
if not dataset_downloaded_successfully:
    print("\n--- Dataset not successfully downloaded. Skipping preprocessing, model building, and training. ---")
else:
    # --- 3. Data Augmentation and Generators ---
    print("Setting up data augmentation and image data generators...")

    # Data augmentation for training data
    train_datagen = ImageDataGenerator(
        rotation_range=15,
        width_shift_range=0.15,
        height_shift_range=0.15,
        shear_range=0.15,
        zoom_range=0.15,
        horizontal_flip=True,
        fill_mode='nearest',
        rescale=1./255 # Normalize pixel values
    )

    # No augmentation for validation and test sets, only rescaling
    val_test_datagen = ImageDataGenerator(rescale=1./255)

    # Create generators from directories
    # Note: For FER2013, the 'test' folder contains both PublicTest and PrivateTest
    # We will use it as a single test set for simplicity, or you can manually split if needed.
    train_generator = train_datagen.flow_from_directory(
        TRAIN_DIR,
        target_size=(IMG_WIDTH, IMG_HEIGHT),
        color_mode='rgb', # VGG16 expects 3 channels
        batch_size=64,
        class_mode='categorical',
        shuffle=True
    )

    # For validation, FER2013 dataset usually has 'PublicTest' as part of the CSV.
    # Since this dataset structure has only 'train' and 'test' folders,
    # we'll use a portion of the 'train' data for validation, or assume 'test' serves as validation/test.
    # Given the user's output, the 'test' directory likely contains all test images.
    # A common practice is to use a validation split during training with the train_datagen,
    # but since the dataset is already split into train/test directories, we'll use the 'test' for evaluation.
    # For a true validation set, you might need to manually split the 'train' directory or find a different dataset.
    # For now, we'll just use the 'test' directory for both validation and final evaluation.
    # If the 'test' directory is large, you might want to create a validation generator from it.

    # Let's create a validation generator from the 'test' directory for monitoring during training
    # and a separate evaluation generator.
    # This dataset (msambare/fer2013) has 'train' and 'test' folders.
    # The 'test' folder contains both public and private test data.
    # For simplicity and to avoid manual splitting, we will use the 'test' directory for evaluation.
    # If you need a separate validation set, you would typically split the 'train' directory or use a different dataset.

    # For this specific dataset, the 'test' folder *is* the combined PublicTest and PrivateTest.
    # We'll use it for evaluation directly.
    # For a validation set during training, we can use a validation_split on the train_datagen,
    # but the problem is that flow_from_directory for validation_split only works on a single directory.
    # Given the structure, we'll train on 'train' and evaluate on 'test'.
    # If you want a validation set, you'd typically split the 'train' data yourself into train/val folders.
    # For simplicity, let's just use the 'test' directory as the validation source for callbacks.

    val_generator = val_test_datagen.flow_from_directory(
        TEST_DIR, # Using TEST_DIR for validation purposes during training
        target_size=(IMG_WIDTH, IMG_HEIGHT),
        color_mode='rgb',
        batch_size=64,
        class_mode='categorical',
        shuffle=False # No need to shuffle validation data
    )

    test_generator = val_test_datagen.flow_from_directory(
        TEST_DIR, # Using TEST_DIR for final evaluation
        target_size=(IMG_WIDTH, IMG_HEIGHT),
        color_mode='rgb',
        batch_size=64,
        class_mode='categorical',
        shuffle=False # No need to shuffle test data
    )

    # Map class indices to emotion labels
    # The class_indices attribute of the generator maps folder names to numerical labels
    # We need to reverse this to get emotion labels from numerical predictions
    class_indices = train_generator.class_indices
    EMOTION_LABELS_REVERSE = dict((v, k) for k, v in class_indices.items())
    print("Emotion labels mapped:", EMOTION_LABELS_REVERSE)

    # Calculate class weights from the training generator's class counts
    # This is important for imbalanced datasets like FER2013
    class_counts = Counter(train_generator.classes)
    total_samples = sum(class_counts.values())
    class_weights = {i: total_samples / (NUM_CLASSES * count) for i, count in class_counts.items()}
    print("Class weights for handling imbalance:", class_weights)


    # --- 4. Model Building with Transfer Learning (VGG16) ---

    print("Building model with VGG16 transfer learning...")

    # Load VGG16 base model, excluding the top (fully connected) layers
    # Input shape must be (IMG_WIDTH, IMG_HEIGHT, 3) for VGG16
    base_model = VGG16(weights='imagenet', include_top=False, input_shape=(IMG_WIDTH, IMG_HEIGHT, 3))

    # Freeze the convolutional base to train only the new layers first
    for layer in base_model.layers:
        layer.trainable = False

    # Add custom classification layers on top of VGG16
    x = base_model.output
    x = Flatten()(x)
    x = Dense(256, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    x = Dense(128, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    predictions = Dense(NUM_CLASSES, activation='softmax')(x)

    model = Model(inputs=base_model.input, outputs=predictions)

    # Compile the model for the first phase of training (feature extraction)
    model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

    model.summary()

    # --- 5. Training Phase 1: Feature Extraction ---

    print("\n--- Training Phase 1: Feature Extraction (Training new layers) ---")

    # Callbacks for training
    checkpoint_phase1 = ModelCheckpoint(
        filepath=MODEL_SAVE_PATH,
        monitor='val_accuracy',
        save_best_only=True,
        mode='max',
        verbose=1
    )
    early_stopping_phase1 = EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True,
        verbose=1
    )
    reduce_lr_phase1 = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,
        patience=5,
        min_lr=0.00001,
        verbose=1
    )

    history_phase1 = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // train_generator.batch_size,
        epochs=20, # Start with a moderate number of epochs
        validation_data=val_generator,
        validation_steps=val_generator.samples // val_generator.batch_size,
        callbacks=[checkpoint_phase1, early_stopping_phase1, reduce_lr_phase1],
        class_weight=class_weights, # Apply class weights
        verbose=1
    )

    # Load the best model from Phase 1 before fine-tuning
    model.load_weights(MODEL_SAVE_PATH)

    # --- 6. Training Phase 2: Fine-tuning ---

    print("\n--- Training Phase 2: Fine-tuning (Unfreezing some base layers) ---")

    # Unfreeze the last few convolutional blocks of VGG16
    # You can experiment with how many layers to unfreeze
    for layer in base_model.layers[-4:]: # Unfreeze last 4 layers of VGG16
        layer.trainable = True

    # Recompile the model with a much lower learning rate for fine-tuning
    model.compile(optimizer=Adam(learning_rate=0.00001), loss='categorical_crossentropy', metrics=['accuracy'])

    model.summary()

    # Callbacks for fine-tuning
    checkpoint_phase2 = ModelCheckpoint(
        filepath=MODEL_SAVE_PATH, # Overwrite with potentially better model
        monitor='val_accuracy',
        save_best_only=True,
        mode='max',
        verbose=1
    )
    early_stopping_phase2 = EarlyStopping(
        monitor='val_loss',
        patience=15, # More patience for fine-tuning
        restore_best_weights=True,
        verbose=1
    )
    reduce_lr_phase2 = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,
        patience=7,
        min_lr=0.000001,
        verbose=1
    )

    history_phase2 = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // train_generator.batch_size,
        epochs=50, # More epochs for fine-tuning
        validation_data=val_generator,
        validation_steps=val_generator.samples // val_generator.batch_size,
        callbacks=[checkpoint_phase2, early_stopping_phase2, reduce_lr_phase2],
        class_weight=class_weights, # Apply class weights
        verbose=1
    )

    # Load the best model after fine-tuning
    model.load_weights(MODEL_SAVE_PATH)

    # --- 7. Evaluation ---

    print("\n--- Evaluating the model on the test set ---")

    test_loss, test_accuracy = model.evaluate(
        test_generator,
        steps=test_generator.samples // test_generator.batch_size,
        verbose=1
    )

    print(f"\nTest Loss: {test_loss:.4f}")
    print(f"Test Accuracy: {test_accuracy:.4f}")

    # Generate predictions for classification report and confusion matrix
    # Need to reset generator before prediction to ensure order
    test_generator.reset()
    y_pred_probs = model.predict(test_generator, steps=test_generator.samples // test_generator.batch_size + 1) # +1 to ensure all samples are covered
    y_pred = np.argmax(y_pred_probs, axis=1)

    # Get true labels from the generator
    y_true = test_generator.classes[:len(y_pred)] # Ensure y_true matches length of y_pred

    # Get target names in the correct order
    target_names = [EMOTION_LABELS_REVERSE[i] for i in range(NUM_CLASSES)]

    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=target_names))

    print("\nConfusion Matrix:")
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix')
    plt.show()

    # Plot training history
    def plot_training_history(history1, history2):
        plt.figure(figsize=(12, 5))

        # Plot accuracy
        plt.subplot(1, 2, 1)
        plt.plot(history1.history['accuracy'], label='Train Accuracy Phase 1')
        plt.plot(history1.history['val_accuracy'], label='Val Accuracy Phase 1')
        # Adjust x-axis for phase 2 to continue from phase 1
        epochs_phase1 = len(history1.epoch)
        epochs_phase2 = len(history2.epoch)
        x_axis_phase2 = np.arange(epochs_phase1, epochs_phase1 + epochs_phase2)

        plt.plot(x_axis_phase2, history2.history['accuracy'], label='Train Accuracy Phase 2')
        plt.plot(x_axis_phase2, history2.history['val_accuracy'], label='Val Accuracy Phase 2')
        plt.title('Model Accuracy')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.grid(True)

        # Plot loss
        plt.subplot(1, 2, 2)
        plt.plot(history1.history['loss'], label='Train Loss Phase 1')
        plt.plot(history1.history['val_loss'], label='Val Loss Phase 1')
        plt.plot(x_axis_phase2, history2.history['loss'], label='Train Loss Phase 2')
        plt.plot(x_axis_phase2, history2.history['val_loss'], label='Val Loss Phase 2')
        plt.title('Model Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True)

        plt.tight_layout()
        plt.show()

    plot_training_history(history_phase1, history_phase2)

    print(f"\nModel saved to: {MODEL_SAVE_PATH}")
    print("Congratulations! You've built and evaluated your emotion detection model.")
    print("The next step is to deploy this model using Streamlit.")

```
